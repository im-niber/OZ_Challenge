[문제2]

Python의 얕은 복사와 깊은 복사에 대해서 설명하세요

얕은 복사는 새로운 객체를 만들고 원본 객체를 가리키는 참조를 (가능한 범위까지) 새로운 객체에 삽입합니다.

깊은 복사는 새로운 객체를 만들고 재귀적으로 원본 객체의 사본을 새로 만든 객체에 삽입합니다. 또 깊은복사는 얕은 복사와 다르게 객체 내부의 객체들까지 모두 새롭게 copy 됩니다.

Python의 Set에 대해 설명해주세요

[문제5]

Python의 list는 연결리스트(linked list) 인가요?

아닙니다. 파이썬에서 list는 PyistObject라는 C언어의 구조체로 이루어져있고 내부 ob_item이 리스트 원소에 관한 포인터 배열로 구현되어있습니다. 이는 리스트 원소들의 주소 값을 저장한 C언어의 배열을 가리킵니다.

조회를 할 때 이중 포인터를 거치게 되므로 C언어의 배열보다 속도가 느립니다.

[문제6]

딕셔너리 자료형에 대해 설명해주세요

해시테이블로 만들어진 자료구조로, 삽입 삭제 탐색시에 O(1)의 시간복잡도를 가집니다. 하지만 해쉬충돌이 일어난다면 O(N)까지 느려질 가능성도 있습니다.
