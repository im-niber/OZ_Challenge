[문제4]

클래스 다중 상속과 관련된 MRO에 대해 설명해주세요

MRO(Method Resolution Order)란 파이썬에서 다중 상속시 메서드 호출 순서를 정의하는 알고리즘입니다.

다중 상속은 하나의 클래스가 여러 개의 클래스로부터 상속받는 것을 의미합니다.

MRO의 규칙은 다음과 같습니다.

1. 탐색순서:
   - 자식클래스부터 시작하여 부모 클래스를 재귀적으로 탐색
   - 왼쪽에서 오른쪽 순서로 상속 목록에 나열된 클래스를 탐색합니다.
   - 이미 탐색된 클래스는 다시 탐색 x
2. 동일 메서드 존재:
   - 탐색 순서에 따라 처음 발견된 메서드가 호출됩니다
   - 동일 메서드가 여러개면 오류가발생
3. 예외:
   - super() 함수를 사용하여 부모 클래스의 메서드를 직접 호출 가능합니다.
     - 예외케이스: https://tibetsandfox.tistory.com/26 의 super 클래스와 MRO 예제코드
   - __mro__속성을 사용하여 MRO 순서를 직접 확인할 수 있습니다.


[문제5]

Self는 어떤 키워드 인가요?

Self는 인스턴스 메서드 내에서 현재 객체를 참조할 때 사용합니다.

클래스 메서드 내에서는 클래스 자체를 참조합니다.

파이썬에서 메소드 첫 번째 인자는 항상 self가 들어갑니다.

[문제6]

ZeroDivisionError,  IndexError 외에 3가지 이상의 파이썬 에러에 대해 설명해주세요

1. KeyError: 딕셔너리 사용시에 없는 Key 값에 접근시 발생
2. ValueError: 부적절한 값을 가진 인자를 받았을 때 발생하는 에러
3. TypeError: 서로 다른 타입으로 연산하려고 할 때 발생
